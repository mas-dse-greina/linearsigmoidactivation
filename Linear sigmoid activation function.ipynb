```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
import math

def sigmoid(x):
  return 1 / (1 + math.exp(-x))
```

```python
t = np.linspace(-10, 10, 10000)
```

```python
sig = []
for i in t:
    sig.append(sigmoid(i))
```

```python
plt.plot(t, sig+0.05*t-0.5);
plt.title("Function doesn't saturate", fontsize=14);
plt.grid();
```

```python
plt.plot(t[1:],np.diff(sig+0.05*t-0.5));
plt.title('Continuous derivative is still like sigmoid', fontsize=14);
```
